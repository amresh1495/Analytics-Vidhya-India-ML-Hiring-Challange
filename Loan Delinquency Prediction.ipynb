```python
# creating dummy variables for object type columns 

X_train = pd.get_dummies(X_train, drop_first=True)
```

```python
X_train.head()
```

```python
# models 

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import BernoulliNB
from sklearn.neighbors import KNeighborsClassifier

nb_clf = BernoulliNB(alpha=1.2, binarize=0.0)
svc_clf = SVC(kernel='linear')
xgb_clf = XGBClassifier()
gb_clf = GradientBoostingClassifier(learning_rate=0.2, loss='exponential', n_estimators=25)
bag_clf = BaggingClassifier()
ada_clf = AdaBoostClassifier()
rf_clf = RandomForestClassifier()
lr_clf = LogisticRegression()
knn_clf = KNeighborsClassifier()
```

```python
# Selecting the best features using Recursive Feature Elimination 

from sklearn.feature_selection import RFE

selector = RFE(rf_clf)
selector = selector.fit(X_train, y_train)
selector.ranking_
temp = pd.Series(selector.support_,index = X_train.columns)
selected_features_rfe = temp[temp==True].index
print(selected_features_rfe)
```

```python
# Dropping the least important features 

X_train = X_train.drop(columns=['number_of_borrowers', 'insurance_type', 'm1', 'm2', 'm3', 'source_Z', 'financial_institution_Chapman-Mcmahon', 'financial_institution_Cole, Brooks and Vincent', 'financial_institution_Edwards-Hoffman', 'financial_institution_Martinez, Duffy and Bird', 'financial_institution_Miller, Mcclure and Allen', 
                      'financial_institution_Nicholson Group', 'financial_institution_Richards-Walters', 
                      'financial_institution_Richardson Ltd', 'financial_institution_Romero, Woods and Johnson', 
                      'financial_institution_Sanchez, Hays and Wilkerson', 'financial_institution_Sanchez-Robinson', 
                      'financial_institution_Suarez Inc', 'financial_institution_Swanson, Newton and Miller',
       'financial_institution_Taylor, Hunt and Rodriguez', 'financial_institution_Thornton-Davis', 
                      'financial_institution_Turner, Baldwin and Rhodes'
                     ])
```

```python
# Trying different techniques to handle imbalanced dataset 

# Build model with no sampling
pip_orig = make_pipeline(gb_clf)

scores = cross_val_score(pip_orig,
                         X_train, y_train,
                         scoring="f1", cv=10)
print(f"Original model's average f1 score: {scores.mean():.5f}")


# Build model with undersampling
pip_undersample = imb_make_pipeline(
                                    RandomUnderSampler(), gb_clf)
scores = cross_val_score(pip_undersample,
                         X_train, y_train,
                         scoring="f1", cv=10)
print(f"Under-sampled model's average f1 score: {scores.mean():.5f}")


# Build model with oversampling
pip_oversample = imb_make_pipeline(
                                    RandomOverSampler(), gb_clf)
scores = cross_val_score(pip_oversample,
                         X_train, y_train,
                         scoring="f1", cv=10)
print(f"Over-sampled model's average f1 score: {scores.mean():.5f}")

# Build model with EasyEnsemble
resampled_rf = BalancedBaggingClassifier(base_estimator=gb_clf,
                                         n_estimators=10, random_state=123)
pip_resampled = make_pipeline(resampled_rf)


# predictions = pip_resampled.fit(X_train, y_train)                             
scores = cross_val_score(pip_resampled,
                         X_train, y_train,
                         scoring="f1", cv=10)
print(f"EasyEnsemble model's average f1 score: {scores.mean():.5f}")


# Build model with SMOTE
pip_smote = imb_make_pipeline(
                              SMOTE(), gb_clf)
predictions = pip_smote.fit(X_train, y_train)
scores = cross_val_score(pip_smote,
                         X_train, y_train,
                         scoring="f1", cv=10)
print(f"SMOTE model's average f1 score: {scores.mean():.5f}")
```

```python
# Doing prediction on the test data provided 

X_test = pd.read_csv('/home/amresh/Desktop/datasets/loan_prediction_ml_hiring_challenge/test.csv')
loan_id = X_test.loan_id
X_test = X_test.drop(columns=['loan_id'])

object_columns_test = list(X_test.select_dtypes(include=['object']).columns)
X_test = pd.get_dummies(X_test, columns=object_columns_test, drop_first=True)
X_test = X_test.drop(columns=['number_of_borrowers', 'insurance_type', 'm1', 'm2', 'm3', 'source_Z', 'financial_institution_Chapman-Mcmahon', 'financial_institution_Cole, Brooks and Vincent', 'financial_institution_Edwards-Hoffman', 'financial_institution_Martinez, Duffy and Bird', 'financial_institution_Miller, Mcclure and Allen', 
                      'financial_institution_Nicholson Group', 'financial_institution_Richards-Walters', 
                      'financial_institution_Richardson Ltd', 'financial_institution_Romero, Woods and Johnson', 
                      'financial_institution_Sanchez, Hays and Wilkerson', 'financial_institution_Sanchez-Robinson', 
                      'financial_institution_Suarez Inc', 'financial_institution_Swanson, Newton and Miller',
       'financial_institution_Taylor, Hunt and Rodriguez', 'financial_institution_Thornton-Davis', 
                      'financial_institution_Turner, Baldwin and Rhodes'])
predicted = pip_smote.predict(X_test)
```

```python
# Generating the submission file 

submission = pd.DataFrame({'loan_id': loan_id, 'm13': predicted })
submission.to_csv("submission_gbc_smote_grid_search.csv", index=False)
```
